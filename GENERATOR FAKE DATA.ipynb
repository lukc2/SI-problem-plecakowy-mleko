```python
import random
#from collections import Counter do zliczania ilości
import json


pojemnoscSzkalnki = random.uniform(0.1, 0.3)
cenaSzklanki = random.uniform(0.7, 1.2)
pojemnoscButelki = random.uniform(0.5, 2)
cenaButelki = random.uniform(0.9, 2)
pojemnoscKartonu = random.uniform(1, 3)
cenaKartonu = random.uniform(0.5, 2.8)
pojemnoscBaniaka = random.uniform(5, 30)
cenaBaniaka = random.uniform(15, 60)
pojemnoscKraty= random.randint(0, 10)*pojemnoscButelki
cenaKraty = pojemnoscKraty * cenaButelki * 0.5
pojemnoscBeczki = random.uniform(30, 120)
cenaBeczki = pojemnoscBeczki * random.uniform(0.9, 3)

cenaZwyklego = random.uniform(0.9, 1.1)
cenaKokosowego = random.uniform(1, 1.3)
cenaCzekoladowego = random.uniform(1.2, 1.6)
cenaTruskawkowego = random.uniform(0.8, 1.9)
cenaWaniliowego = random.uniform(0.8, 1.1)
cenaPistacjowego = random.uniform(1.3, 1.7)

cenaKrowiego=random.uniform(0.9,1.1)
cenaKoziego=random.uniform(0.8,1)
cenaOwczego=random.uniform(0.6,0.8)
cenaWielbladziego=random.uniform(0.7,0.9)
cenaSojowego=random.uniform(2.0,2.9)
cenaRyzowego=random.uniform(2.2,2.5)
cenaLudzkiego=random.uniform(3.0,5.0)

def pojemnik(nazwa, pojemnosc, cena):
    rand = random.randint(0, 100)
    
    if rand<15:
        nazwa+="szlanka mleka"
        pojemnosc = pojemnoscSzkalnki
        cena = cenaSzklanki
    elif rand<35:
        nazwa+="butelka mleka"
        pojemnosc = pojemnoscButelki
        cena = cenaButelki
    elif rand<65:
        nazwa+="karton mleka"
        pojemnosc = pojemnoscKartonu
        cena = cenaKartonu
    elif rand<75:
        nazwa+="baniak mleka"
        pojemnosc = pojemnoscBaniaka
        cena = cenaBaniaka
    elif rand<85:
        nazwa+="krata mleka"
        pojemnosc = pojemnoscKraty
        cena = cenaKraty
    else:
        nazwa+="beczka mleka"
        pojemnosc = pojemnoscBeczki
        cena = cenaBeczki
        
    return nazwa, pojemnosc, cena

def smak(nazwa, cena):
    rand = random.randint(0, 100)
    
    if rand<50:
        nazwa+=" zwyklego"
        cena *= cenaZwyklego
    elif rand<60:
        nazwa+=" kokosowego"
        cena *= cenaKokosowego
    elif rand<70:
        nazwa+=" czekoladowego"
        cena *= cenaCzekoladowego
    elif rand<80:
        nazwa+=" truskawkowego"
        cena *= cenaTruskawkowego
    elif rand<90:
        nazwa+=" waniliowego"
        cena *= cenaWaniliowego
    else:
        nazwa+=" pistacjowego"
        cena *= cenaPistacjowego

    return nazwa, cena

def zwierze(nazwa,cena):
    rand = random.randint(0, 100)
    if rand<50:
        nazwa+=" krowiego"
        cena*=cenaKrowiego
    elif rand<60:
        nazwa+=" sojowego"
        cena*=cenaKoziego
    elif rand<70:
        nazwa+=" ryzowego"
        cena*=cenaRyzowego
    elif rand<80:
        nazwa+=" koziego"
        cena*=cenaKoziego
    elif rand<90:
        nazwa+=" owczego"
        cena*=cenaOwczego
    elif rand<99:
        nazwa+=" wielbladziego"
        cena*=cenaWielbladziego
    else:
        nazwa+=" ludzkiego"
        cena*=cenaLudzkiego
    return nazwa, cena

def procent(nazwa,cena):
    rand = random.randint(0, 100)
    if rand<10:
        nazwa+=" 0%"
        cena-=0.4
    elif rand<40:
        nazwa+=" 1%"
        cena-=0.2
    elif rand<60:
        nazwa+=" 1.5%"
        cena+=0.1
    elif rand<90:
        nazwa+=" 2%"
        cena+=0.3
    else:
        nazwa+=" 3.2%"
        cena+=0.4

    return nazwa, cena

def pochodzenie(nazwa,cena):
    rand = random.randint(0, 100)
    if rand<30:
        nazwa+=" prosto z farmy"
        cena=abs((cena**2)-cena*2)%1000
    elif rand<90:
        nazwa+=" z wielkiej hodowli"
        cena=cena
    else:
        nazwa+=" z proszku"
        cena=abs(cena*2-cena**2)%500

    return nazwa, cena


def jakosc(nazwa,cena):
    rand = random.randint(0, 100)
    if rand<15:
        nazwa+=" zlej jakosci"
        cena*=0.5
    elif rand<40:
        nazwa+=" niedobrej jakosci"
        cena*=0.75
    elif rand<70:
        nazwa+=" sredniej jakosci"
        cena*=1
    elif rand<85:
        nazwa+=" dobrej jakosci"
        cena*=1.1
    elif rand<95:
        nazwa+=" bardzo dobrej jakosci"
        cena*=1.2
    else:
        nazwa+=" znakomitej jakosci"
        cena*=1.3
    return nazwa, cena

def pojedynczeMleko():
    nazwa = ""
    pojemnosc = 0.0
    cena = 0.0
    
    nazwa, pojemnosc, cena = pojemnik(nazwa, pojemnosc, cena)
    nazwa, cena = smak(nazwa, cena)
    nazwa, cena = zwierze(nazwa, cena)
    nazwa, cena = procent(nazwa, cena)
    nazwa, cena = pochodzenie(nazwa, cena)
    nazwa, cena = jakosc(nazwa, cena)
    
    return {"name": nazwa, "volume": round(pojemnosc,2), "price": round(cena,2)}

tablica = []

for i in range(0,20000):
    tablica.append(pojedynczeMleko())
    
#for i in tablica:
#    print(i, '\n')


with open('mleko.json', 'w') as json_file:    
    json.dump(tablica, json_file, indent=4)
    

```
