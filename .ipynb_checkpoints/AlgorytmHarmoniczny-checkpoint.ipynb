```python
import json
import random

with open('mleko.json') as f:
    data = json.load(f)
    

HMS = 20


HM = []
koszty = []
fitness = []

def get_item(szukany):
    for item  in data: 
        if item['id'] == szukany:
            return item

for i in range(0, HMS):
    pojemnosc = 200
    przepelniony = False
    koszt = 0
    przystosowanie = 0
    wektor = []
    
    while(not przepelniony):
        randId = random.randint(0, 19999)
        while(randId in wektor):
            randId = random.randint(0, 19999)
            
        item = get_item(randId)
        
        if(pojemnosc - item["volume"] < 0):
            przepelniony = True
        else:
            pojemnosc -= item["volume"]
            wektor.append(randId)
            koszt += item["price"]
            
    przystosowanie = 0.6 * pojemnosc + 0.4 * koszt 
    HM.append(wektor)
    koszty.append(round(koszt,2))
    fitness.append(round(przystosowanie,2))
    
    
print("PAMIEC HM")

for i in HM:
    print(i)
print("KOSZTY")
print(koszty)
print("Fitness")
print(fitness)
```

```python
HMCR = 60
COLUMN_COUNT = 0

for i in HM:
    if(len(i) > COLUMN_COUNT):
        COLUMN_COUNT = len(i)
        
print(COLUMN_COUNT)


for i in range(0, 2000): 
    print('iteracja', i)
    r1 = random.randint(0, 100)
    pojemnosc = 200
    przepelniony = False
    wektor = []
    koszt = 0
    przystosowanie = 0
    
    
    
    if(r1< HMCR):
        for j in range (0, COLUMN_COUNT):
            if(przepelniony): break
                
                
            row = random.randint(0, 19)
            while(len(HM[row])-1 < j or HM[row][j] in wektor): 
                row = random.randint(0, 19)
                print('len', len(HM[row])-1, 'kolumna', j)
            
            item = get_item(HM[row][j])

            if(pojemnosc - item["volume"] < 0):
                    przepelniony = True
            else:
                pojemnosc -= item["volume"]
                wektor.append(HM[row][j])
                koszt += item["price"]
            
                    
    else:
        while(not przepelniony):
            randId = random.randint(0, 19999)
            while(randId in wektor):
                randId = random.randint(0, 19999)   
            item = get_item(randId)
            if(pojemnosc - item["volume"] < 0):
                przepelniony = True
            else:
                pojemnosc -= item["volume"]
                wektor.append(randId)
                koszt += item["price"]
    
    
    przystosowanie = 0.6 * pojemnosc + 0.4 * koszt 
    przystosowanie = round(przystosowanie, 2)
    
    
    
    
    
    
    for i in range(0,len(fitness)):
        if(fitness[i] > przystosowanie): 
            fitness[i] = przystosowanie
            koszty[i] = koszt
            HM[i] = wektor
            break
            
            
            
            
            
                
for i in HM:
    print(i)       
print("Fitness")
print(fitness)
print(koszty)
            
        
```
